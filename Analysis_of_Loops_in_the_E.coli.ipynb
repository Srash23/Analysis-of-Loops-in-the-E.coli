```python
import networkx as nx
import matplotlib.pyplot as plt
import random

# Set the random seed for reproducibility
random.seed(42)

# Create a directed graph with 197 nodes and 477 edges
g = nx.gnm_random_graph(197, 477, directed=True)

# Add self-loops
for _ in range(477):
    n = random.choice(list(g.nodes))
    g.add_edge(n, n)

# Plotting the graph
plt.figure(figsize=(12, 12))
pos = nx.spring_layout(g)  # Layout for the nodes
nx.draw(g, pos, with_labels=False, node_size=50, arrows=True, arrowstyle='-|>', arrowsize=10)

plt.show()

# Counting nodes, edges, and self-loops
num_nodes = g.number_of_nodes()
num_edges = g.number_of_edges()
num_self_loops = sum(1 for n in g.nodes if g.has_edge(n, n))

print("Number of nodes: ", num_nodes)
print("Number of edges: ", num_edges)
print("Number of self-loops: ", num_self_loops)

```
